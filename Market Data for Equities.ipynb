#### Import Liabires

```python
import pandas as pd
import yfinance as yf
from datetime import date
```

```python
date.today()
```

#### Equity Spot Prices

```python
spotData = yf.download(tickers = 'HDFCBANK.NS', start = date(2020,1,1), end = date.today(), 
                       auto_adjust = False, multi_level_index = False)
spotData
```

```python
spotData.loc[:,'Adj Close']
```

```python
spotData['Adj Close'].plot(figsize = (14,6), legend = True)
```

```python
spotData = yf.download(tickers = 'HDFCBANK.NS', start = date(2025,3,10), end = date.today(), interval = '1m',
                       auto_adjust = False, multi_level_index = False)
spotData
```

```python
stockTickers = ['ICICIBANK.NS','AXISBANK.NS']
spotData = yf.download(tickers = stockTickers, start = date(2020,1,1), end = date.today(),
                       auto_adjust = False, multi_level_index = False)
spotData
```

```python
spotData['Adj Close']['ICICIBANK.NS']
```

```python
spotData['Volume']['AXISBANK.NS']
```

```python
spotData['Adj Close'].plot(figsize = (14,6), legend = True)
```
