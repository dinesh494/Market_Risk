### 1. Normal Distrubution

#### <code> Standard Normal Distrubution<code>

```python
normalDist = stats.norm()  #defalut loc=0, scale=1
```

```python
print(normalDist.mean(),normalDist.std())
```

```python
print(np.random.normal())
```

```python
print(normalDist.rvs(size=10))
```

```python
np.random.seed(0)
print(normalDist.rvs(size=10))
```

```python
randomNum = normalDist.rvs(size=1000000)
print(randomNum)
```

```python
fig, (probDist, cumDist) = plt.subplots(1, 2, figsize=(14,6))

probDist.hist(randomNum, bins=250)
probDist.set_title('Probability Distribution')

cumDist.hist(randomNum, bins=250, cumulative=True)
cumDist.set_title("Cumualtive Probaility Distubution")

plt.show()
```

#### <Code>Probability Density Function (PDF)<Code>

```python
print(normalDist.pdf(0))
```

```python
print(normalDist.pdf(3))
```

```python
print(normalDist.pdf(-3))
```

```python
randomNumLine = np.linspace(start=randomNum.min(),stop=randomNum.max())
print(randomNumLine)
```

```python
randomNumProb = normalDist.pdf(randomNumLine)
print(randomNumProb.round(2))
```

```python
print(tuple(zip(randomNumLine,randomNumProb.round(4))))
```

#### <Code>Cumulative Distribution Function (CDF)<Code>

```python
normalDist.cdf(0)
```

```python
normalDist.cdf(3)
```

```python
print(normalDist.cdf(3) - normalDist.cdf(-3))
print(normalDist.cdf(2) - normalDist.cdf(-2))
print(normalDist.cdf(1) - normalDist.cdf(-1))
```

#### `Precent Point Function (PPF)`

```python
normalDist.ppf(0.5)
```

```python
normalDist.ppf(0.99), normalDist.ppf(0.01)
```

### `Normal Distribution VS Standard Normal Distribution`

```python
print(randomNum.mean(),randomNum.std(),stats.skew(randomNum), stats.kurtosis(randomNum))
```

```python
stats.norm(loc=randomNum.mean(), scale=randomNum.std())
```

```python
stats.norm.ppf(1-0.99, randomNum.mean(), randomNum.std())
```
