## `Normal Distribution VS Log-Normal Distribution`

    Portfolio of 1000000 Stocks: PCA
    
    Normal Distribution: Mean, Std
        Mean of 10 Stocks (10 Parameter Values), Std of 10 Stocks (10 Parameter Values)
        Total (20 Parameter Values)
    
    Log-Normal Distribution: Mean, Std, Skewness, Kurtosis
        Mean of 10 Stocks (10 Parameter Values), Std of 10 Stocks (10 Parameter Values), Skewness of 10 Stocks (10 Parameter Values), Kurtosis of 10 Stocks (10 Parameter Values)
        Total (40 Parameter Values)
    
        Std formula to calculation portfolio std = Weights, Correlation

```python
normRandomNum = stats.norm().rvs(size=10000)
lognormRandomNum = stats.lognorm(s=1).rvs(size=10000)
```

```python
transnormRandomNum = np.log(lognormRandomNum)
print(transnormRandomNum)
```

```python
fig, (probDist1,probDist2,probDist3) = plt.subplots(1,3,figsize = (14,5))

probDist1.hist(normRandomNum, bins=250, label="Normal Distribution")
probDist1.set_title('Probability Distribution')
probDist1.legend()

probDist2.hist(lognormRandomNum, bins=250, label="Log-Normal Distribution")
probDist2.set_title('Probability Distribution')
probDist2.legend()

probDist3.hist(transnormRandomNum, bins=250, label="Transformed Log-Normal Distribution")
probDist3.set_title('Probability Distribution')
probDist3.legend()

plt.show()
```

```python
mean = np.exp(transnormRandomNum.mean() + 0.5 * transnormRandomNum.std()**2)
var = (np.exp(transnormRandomNum.std()**2) - 1)*np.exp(2*transnormRandomNum.mean() +transnormRandomNum.std())
std = np.sqrt(var)
skew = np.sqrt(np.exp(transnormRandomNum.std()**2)-1)*(np.exp(transnormRandomNum.std()**2)+2)
kurt = 1 * np.exp(4*transnormRandomNum.std()**2)+2 * np.exp(3*transnormRandomNum.std()**2)+3 * np.exp(2*transnormRandomNum.std()**2)-6
print(mean,var,std,skew,kurt)
```

```python
skew = stats.lognorm.stats(s=1,moments='s')
kurt = stats.lognorm.stats(s=1,moments='k')
print(skew,kurt)
```
