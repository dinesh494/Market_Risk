```python
CCYs_ = []
for Currencies in CCYs.values():
    for ccode  in Currencies:
        CCYs_.append(ccode + '=X')

print(CCYs_)
```

```python
CCYs_ = [ccode + '=X' for currencies in CCYs.values() for ccode in currencies]
print(CCYs_)
```

```python
spotData = yf.download(tickers=CCYs_, period='10y', auto_adjust=False)['Adj Close']
spotData
```

#### FX Rate Conversion

```python
spotData[['AUD=X', 'CHF=X', 'EUR=X', 'INR=X', 'JPY=X']].tail(3).head(1)
```

```python
USDEURrate= spotData['EUR=X'].tail(3).head(1).values[0]
print(USDEURrate)
```

             INR/EUR = INR/USD * USD/EUR, where USD is the priceCurrency and EUR is the baseCurrency.
             Or, INR/EUR = INR/USD / EUR/USD

```python
EURUSDrate = 1 / spotData['EUR=X'].tail(3).head(1).values[0]
print(EURUSDrate)
```

```python
spotData[['AUD=X', 'CHF=X', 'EUR=X', 'INR=X', 'JPY=X']].tail(3).head(1) * EURUSDrate
```

```python
spotData[['AUD=X', 'CHF=X', 'EUR=X', 'INR=X', 'JPY=X']].tail(3).head(1) / USDEURrate
```

```python
spotData[['EUR=X' , 'CHF=X']].plot(figsize = (14,6), legend = True)
```
