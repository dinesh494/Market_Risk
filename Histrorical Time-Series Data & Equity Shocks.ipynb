#### Import Libaries

```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import date
```

#### Data Extraction

```python
stockTickers = list(pd.read_csv('https://archives.nseindia.com/content/indices/top10nifty50_130325.csv')['SYMBOL'] + '.NS')
print(stockTickers)
```

```python

spotData = yf.download(tickers=stockTickers, 
                       start=date(2021, 1, 1), 
                       end=date.today(),
                       auto_adjust=True)['Close']
print(spotData)
```

# Equity Returns

## 1. Absolute Returns 

        Formula :> Absolute = Curr. Price - Previous Price
            rAbsolute = St - St-1

```python
spotData.diff().dropna()
```

```python
spotData.diff().iloc[1:]
```

## 2. Discrite Propational Returns

        Formula :>  Discrite Propational Returns = (Currn. Price - Prev. Price) / Prev. Price
            rDiscrite = (St-St-1)/St-1

```python
spotData.pct_change().iloc[1:]
```

## 3. Continous Propational Returns
